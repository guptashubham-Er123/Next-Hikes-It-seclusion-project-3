# Feature Engineering:

```python
housing_data["TotalSF"] = housing_data["GrLivArea"] + housing_data["TotalBsmtSF"]
```

```python
housing_data
```

```python
housing_data["Age"] = housing_data["YrSold"] - housing_data["YearBuilt"]
```

```python
housing_data
```

```python
housing_data["Remodeled"] = np.where(housing_data["YearRemodAdd"] != housing_data["YearBuilt"], 1, 0)
```

```python
housing_data
```

```python
housing_data["TotalBathrooms"] = housing_data["FullBath"] + 0.5 * housing_data["HalfBath"] + housing_data["BsmtFullBath"] + 0.5 * housing_data["BsmtHalfBath"]
```

```python
housing_data
```

```python
housing_data["OverallQualCond"] = housing_data["OverallQual"] + housing_data["OverallCond"]
```

```python
housing_data
```

```python
housing_data["TotalPorchSF"] = housing_data["OpenPorchSF"] + housing_data["EnclosedPorch"] + housing_data["3SsnPorch"] + housing_data["ScreenPorch"]
```

```python
housing_data
```

```python
housing_data["LotFrontageRatio"] = housing_data["LotFrontage"] / housing_data["LotArea"]
```

```python
housing_data
```

```python
housing_data["PricePerSF"] = housing_data["SalePrice"] / housing_data["TotalSF"]
```

```python
housing_data
```

```python
housing_data["TimeSinceRemodel"] = housing_data["YrSold"] - housing_data["YearRemodAdd"]
```

```python
housing_data
```

```python
housing_data["TotalSF"] = housing_data["GrLivArea"] + housing_data["TotalBsmtSF"]
```
